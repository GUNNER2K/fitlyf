```python

df['Date_TS'] =df['Date_TS'].apply(lambda x : x.split('T')[0] + " "+x.split('T')[1]) #changing the format from YYYY-MM-DDTH:M:S to YYYY-MM-DD H:M:S
df['Date_TS'] =df['Date_TS'].apply(lambda x : x.split('.')[0]) #ignoring the decimal place in second's place
df
```

```python
# ANSWER TO 1 a and b
df['Date_TS'] = pd.to_datetime(df['Date_TS']) #converting Date_TS to datetime object
print("min_Date_TS = ",df['Date_TS'].min(),"\n","max_Date_TS =" ,df['Date_TS'].max()) #printing min_Date_TS and max_Date_TS()
print ( " the no. of days in between = ",(df['Date_TS'].max()-df['Date_TS'].min())/np.timedelta64(1,'D')) #subtracting to find the number of days in between
print ( " the no. of hours in between = ",(df['Date_TS'].max()-df['Date_TS'].min())/np.timedelta64(1,'D')*24) #multiplying 24 to above result to find no. of hours in between
```

```python
min_Date_TS =df['Date_TS'].min()
max_Date_TS =df['Date_TS'].max()
```

```python
#ANSWER TO 1 c
lis = [{1:(min_Date_TS,(min_Date_TS+timedelta(hours=12)))}] #initializing the first tuple 
i=2
newdate = min_Date_TS+timedelta(hours=12) #new date is initialized to (min_Date_TS+12 hrs)
while(True):
    if((newdate+timedelta(hours=12)) <= max_Date_TS):
        dic = {i:(newdate,newdate+timedelta(hours=12))}# i contains the batch number , newdate contains the current date and (newdate+timedelta(hours=12)) contains date after 12 hours
        lis.append(dic)
        newdate = newdate+timedelta(hours=12) #updating newdate for next batch
        i=i+1 #updating batch number
    else:
        break
lis
# 15 batches are created , 15*12 = 180 hours  , so extra 8 hours remain , we need to accomodate that
dic = {16:(newdate,max_Date_TS)}
lis.append(dic)
lis
```

```python
#ANSWER TO 1 d
# creating a function which accepts datetime object as argument to get batch number
def get_batch(date):
    i=0
    while(i<len(lis)):
        if date >=lis[i][i+1][0] and date<=lis[i][i+1][1]: # let me break it down ,for i=0, lis[i] will return {1: (Timestamp('2019-04-26 00:00:00'), Timestamp('2019-04-26 12:00:00'))},now i+1 is 1 , so lis[i][i+1] will give (Timestamp('2019-04-26 00:00:00'), Timestamp('2019-04-26 12:00:00')) and lis[i][i+1][0] will give only Timestamp('2019-04-26 00:00:00')  
            return i+1
            break
        i=i+1
```

```python
df['Batch'] = df['Date_TS'].apply(lambda x: int(get_batch(x)))
df
```

# SOLVING Q5--

```python
version_data = df[df['Versions']=='T1'] # creating a whole new dataset which only contains Versions==T1 (highest rewarding version)
def get_best_reward_rate(attribute): # Defining a function that takes the column name as an argument and returns the category with highest reward rate along with its value
    reward_dict={}
    for i in version_data[attribute].unique(): #iterating i through every unique value of the column
        reward_rate = len(version_data[(version_data[attribute]==i)&(version_data['reward']==1.0)].Session_IDs.unique())/len(version_data[(version_data[attribute]==i)&(version_data['reward']==0.0)].Session_IDs.unique())# calculating the reward rate of i
        reward_dict[i]=reward_rate # storing the reward rate of i in a dictionary with the key name
    return max(reward_dict) , reward_dict[max(reward_dict)] #returning the maximum value of the dictinary along with key name
```

```python
print("The highest reward rate From T1 among ''Device'' is {} with value {}".format(get_best_reward_rate('Device')[0],get_best_reward_rate('Device')[1]))
print("The highest reward rate From T1 among ''Operating system'' is {} with value {}".format(get_best_reward_rate('Operating system')[0],get_best_reward_rate('Operating system')[1]))
print("The highest reward rate From T1 among ''browser'' is {} with value {}".format(get_best_reward_rate('browser')[0],get_best_reward_rate('browser')[1]))
```
